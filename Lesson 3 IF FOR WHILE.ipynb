### FOR 

```python
numbers=[1,2,4,7,18]

for number in numbers:
    print(number+2)
```

### IF



```python
a=3
b=2

if a<b:
    print('the first one is smaller')
elif a>b:
    print('the first one is bigger')
else:
    print('they must be equal')
    

```

### WHILE

```python
count=0
while count<5:
    print(count,'is less than 5')
    count=count+1
```

### FOR + IF

```python
#go through all the number in the numbers list and check if its module is 0.
for number in numbers:
    if number%2==0:
        print(number,' is an even number')
    else:
        print(number,' is an odd number')
        
```

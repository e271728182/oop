## Lists

https://docs.python.org/3/tutorial/datastructures.html <br>
https://developers.google.com/edu/python/lists <br>
https://www.programiz.com/python-programming/list

```python
#declaring a list via []
my_list=[]
print('my_list is:',my_list)

#declaring a list via keyword list()
my_list_2=list()
print('my_list_2 is:',my_list_2)

#declaring a list with values
my_other_list=[2,1,'a',True]
print('my_other_list is:',my_other_list)

```

### Accessing elements of a list 

List index start at 0 and the last element has index equal to lenght of the list minus 1

```python
print('my_other_list is of lenght:',len(my_other_list))
print('first element is:',my_other_list[0])
print('last element is:',my_other_list[3])
```

### adding elements to a list via append & extend

```python
my_list.clear()
#adding 2 elements
my_list.append(4)
my_list.append(5)

print('my_list is:',my_list)

#extending the list
my_other_list.extend([7,8,9])
print('my_other_list is:',my_other_list)
```

### deleting elements via pop or remove

```python
#remove() will remove a value if it is in the list
my_other_list.remove(2)
print('my_other_list after removing 2:',my_other_list)

#pop() will remove the value at index n
my_other_list.pop(0)
print('my_other_list after removing element at index 0:',my_other_list)
```

### Sorting a list 

```python
new_list=[3,1,9,2,1,5]
new_list.sort()
new_list
```

### Various other interesting properties 

```python
#nested list
nested_list=[['a,b,c'],[1,2,3]]
nested_list
```
